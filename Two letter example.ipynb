# A very simple convolutional net example with a two letter alphabet

In this example a basic examples of convolutions are explained. 

The idea is to give an example to the simple classification of letters in a two letter alphabet. The basic idea is presented by Luis Serrano in this [video](https://www.youtube.com/watch?v=2-Ol7ZB0MmU])


The idea is to show the principles of classification using two letters / and \. Those letters can be written perfectly in a 2x2 grid as.

$\begin{matrix} 
1 & 0 \\
0 & 1 
\end{matrix} \textrm{ and } 
\begin{matrix} 
0 & 1 \\
1 & 0 
\end{matrix}
$


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D
```

```python
# Define the letters
one=np.array([[1.0, 0.0],
              [0.0, 1.0]])

two=np.array([[0.0, 1.0],
              [1.0, 0.0]])

#Define some poorly written examples
test_one = np.array([[1, 0],
                     [0, 0]])

test_two = np.array([[0.1, 0],
                     [1, 0]])

x_train=np.array([one,two])
y_train=np.array([0,1])
x_train = x_train.reshape(-1,2, 2, 1)

x_test=np.array([test_one, test_two])
y_test=np.array([0,1])
x_test = x_test.reshape(-1,2, 2, 1)
```

```python

# Creating a Sequential Model and adding the layers
model = Sequential()
# There is only one filter needed and it should have the same size as 
model.add(Conv2D(1, kernel_size=(2,2), input_shape=(2, 2, 1)))

model.add(Dense(2, activation=tf.nn.softmax))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy','acc'])

history=model.fit(x_train, y_train,epochs=300,batch_size=2)
```

```python
from matplotlib import pyplot as plt
print(model.predict(x_test))
for i in range(1):
    filter1=model.layers[0].get_weights()[0][:,:,0,i]
    wheight = model.layers[0].get_weights()[1][i]
    plt.imshow(filter1,vmin=-0.2,vmax=0.2)
    plt.title(wheight)
    plt.colorbar()
    plt.show()
```
